```octave
# Using for loop

function fibo(n)
    a=0;
    b=1;
    x(:,1)=[1];
    for i=2:n
        c=a+b;
        x(:,i)=[c];
        a=b;
        b=c;
    endfor
    g=sprintf("%d   ",x);
    fprintf("Fibonacci sequence:   %s\n",g)
endfunction    

a = 10;
fibo(a)
```

```octave
# Using while loop
function fibo(n)
    a=0;
    b=1;
    x(:,1)=[1];
    i=2;
    while i<=n
        c=a+b;
        x(:,i)=[c];
        a=b;
        b=c;
        i=i+1;
    endwhile
    g=sprintf("%d   ",x);
    fprintf("Fibonacci sequence:   %s\n",g)
endfunction    

a = 10;
fibo(a)
```
