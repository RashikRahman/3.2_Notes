**Backward Difference Approximation of the First Derivative(BDD)**

This is a backward difference approximation as you are taking a point backward from X. deltaX tends to 0, so eventually after some iterations newX(point) will be backwards of X. We have value of deltaX so as the new point is backward of X then we can say that `new_point` = X - deltaX.

![image.png](figures/fig68.PNG)
![image.png](figures/fig69.PNG)
![image.png](figures/fig70.PNG)

The main difference between FDD and BDD is in FDD the new point is forward of X and in BDD the new point is backward of X.

**`NB` : Run is deltaX and Rise is f(X+deltaX) - f(X)**

**Central Difference Approximation of the First Derivative(CDD)**

As you can see, both forward and backward divided difference approximations of the
first derivative are accurate on the order of O(x). Can we get better approximations? Yes,
another method to approximate the first derivative is called the central difference
approximation of the first derivative.
