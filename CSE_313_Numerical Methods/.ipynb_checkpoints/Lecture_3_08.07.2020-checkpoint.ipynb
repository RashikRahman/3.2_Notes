<p><strong><sup><u>SIGNIFICANT FIGURES</u></sup></strong></p>
<p>RULE #1 -&nbsp;<strong>All digits 1 through 9 are significant</strong><br /> If the mass of an object is measured as 15.8 g, this means that the mass is known to lie between 15.7 and 15.9g. There are 3 significant figures in 15.8. THE INSTRUMENT USED TO MAKE THIS MASS MEASUREMENT CAN ONLY DETECT DIFFERENCES OF 0.1G OR 100 MG.<br /> If the mass of an object is measured as 12.3456 g, this means that the mass is known to lie between 12.3455 and 12.3457 g. There are 6 significant figures in 12.3456. THE INSTRUMENT USED TO MAKE THIS MASS MEASUREMENT CAN ONLY DETECT DIFFERENCES OF 0.1 MG OR 100 MICROGRAMS.<br />Overall, you are counting all of the "certain" digits and a final digit that is uncertain, but nevertheless significant.</p>
<hr width="100%" />
<p><br />RULE #2 -&nbsp;<strong>Zero is significant when it is between two non-zero digits</strong><br /> The quantities 306, 30.6, 3.06 and 0.306 all contain 3 significant figures since the 0 between the 3 and 6 is significant.&nbsp;<u>The number 306 means that the true value rests somewhere between 305 and 307, thus, the zero is known with certainty and is significant</u>. That is, zeros&nbsp;<em>within&nbsp;</em>&nbsp;a number are always significant.</p>
<hr width="100%" />
<p><br />RULE #3 -&nbsp;<strong>Trailing zeros that are not used to hold the zero are not significant</strong><br /> The quantities 279.0, 27.90 and 2.790 all contain 4 significant figures.Again, the first three numbers are known with certainty and the final number is always taken as significant.<br />The quantities 0.2790 and 0.27900 have 4 and 5 significant figures, respectively. In the number 0.27900, it does NOT matter that there are two consecutive zeros. The first zero is known with certainty and the final zero while not known with certainty is still significant. Thus, 4.000 has 4 significant figures.</p>
<hr width="100%" />
<p><br />RULE #4 -&nbsp;<strong>A zero used to fix a decimal point is never significant.</strong><br /> The quantities 0.456, 0.0456 and 0.00456 all contain 3 significant figures. In this case, you need to think in terms of exponential numbers. 0.0456 is 4.56 x 10<sup>-2</sup>&nbsp;(only 3 significant figures) and 0.00456 is 4.56 x 10<sup>-3&nbsp;</sup>(again, only three significant numbers). Thus, 470,000 has only 2 significant figures. However, 470,000 with a line drawn above the final zero or the presence of a decimal point indicate that the measurement had six significant figures. 0.000000004 has only one significant figure; the remaining zeros were used to fix the decimal point.</p>
<hr width="100%" />
<p><br /><strong>Questions</strong>:<br />1.How many significant figures are in the number 0.01020?<br /> Ask yourself: how many figures are known with certainty? Ans. = 4</p>
<p>2. Which of the following has the smallest number of significant figures?0.00030, 123, 0.4005, 2.04, 2.004, 123 and 2.04. Here 0.00030 is the same as 3.0 x 10<sup>-4</sup>, so it has only 2 significant figures.</p>
<p>3. How many significant figures are in the number 20.010?<br />Once again, four digits are known with certainty and there are five significant figures.</p>
<p>4.How many significant figures are in the number 0.01000?<br />This is tricky and requires some thought. As an exponential number it can be represented as 1.000 x 10<sup>-2</sup>&nbsp;and then the presence of FOUR significant figures becomes evident.</p>
<p>5.How many of the following numbers have 4 significant figures? 3.0156, 18.00, 0.007000, 3.45 x 10<sup>4</sup> and 0.0021</p>
