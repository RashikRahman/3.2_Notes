$$f'(x)=\frac{f(x+x)-f(x-x)}{2x}$$

**Here,**  

$$f(x) = e^x$$
$$x = 3$$
$$2x = 2*0.2=0.4$$

**So,**

$$f'(x)=\frac{f(3.2)-f(2.8)}{0.4}$$

$$f'(x)=\frac{e^{3.2}-e^{2.8}}{0.4}$$

$$f'(x)=\frac{24.533-16.445}{0.4}$$

$$f'(x)=20.22$$

True value is 20.22 and approx value is 20.22 so absolute relative true error is = |(20.22-20.22)/20.22| * 100 = 0%
